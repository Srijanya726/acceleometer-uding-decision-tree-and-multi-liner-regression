```python
#importimg decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
#importimg decision tree regressor
from sklearn.tree import DecisionTreeRegressor
```

```python
#creating the decision tree function
dt_model = DecisionTreeRegressor(random_state=10)
```

```python
#fitting the model
dt_model.fit(x_train , y_train)
```

```python
#checking the training score
dt_model.score(x_train, y_train)
```

```python
new_y = []
for i in range(len(y_pred)):
    if y_pred[i] < 0.4:
       new_y.append(0)
    else:
       new_y.append(1)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
train_accuracy = []
validation_accuracy = []
for depth in range(1,10):
    dt_model = DecisionTreeRegressor(max_depth=depth, random_state=10)
    dt_model.fit(x_train,y_train)
    train_accuracy.append(dt_model.score(x_train, y_train))
    validation_accuracy.append(dt_model.score(x_valid, y_valid))
```

```python
frame = pd.DataFrame({'max_depth':range(1,10), 'train_acc':train_accuracy, 'valid_acc':validation_accuracy})
frame.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(frame['max_depth'], frame['train_acc'], marker='o')
plt.plot(frame['max_depth'], frame['valid_acc'], marker='o')
plt.xlabel('Depth of tree')
plt.ylabel('performance')
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression

# create X and y
feature_cols = ['wconfid', 'pctid', 'x','y']
x = data[feature_cols]
y = data.z

# instantiate and fit
l1 = LinearRegression()
l1.fit(x, y)

# print the coefficients
print(l1.intercept_)
print(l1.coef_)
```

```python
l1.score(x,y)  #accuracy#
```

```python
from sklearn.metrics import r2_score
l2 = LinearRegression().fit(x[['wconfid', 'pctid']], y)
l2_preds = l2.predict(x[['wconfid', 'pctid']])

print("R^2: ", r2_score(y, l2_preds))
```

```python
l3 = LinearRegression().fit(x[['wconfid', 'pctid', 'x','y']], y)
l3_preds = l3.predict(x[['wconfid', 'pctid', 'x','y']])

print("R^2: ", r2_score(y, l3_preds))
```

```python
from sklearn.metrics import mean_squared_error
x_valid = data[['wconfid', 'pctid', 'x','y']]
y = data.z
x_train, x_valid, y_train, y_valid= train_test_split(x, y, random_state =0 , test_size=0.3)

l4 = LinearRegression()
l4.fit(x_train, y_train)
l4_preds = l4.predict(x_valid)

print("RMSE :", np.sqrt(mean_squared_error(y_valid, l4_preds)))
print("R^2: ", r2_score(y_valid, l4_preds))
```

```python
x_valid = data[['wconfid', 'pctid', 'x','y']]
y = data.z
x_train, x_valid, y_train, y_valid= train_test_split(x, y, random_state =0 , test_size=0.3)

l5 = LinearRegression()
l5.fit(x_train, y_train)
l5_preds = l4.predict(x_valid)

print("RMSE :", np.sqrt(mean_squared_error(y_valid, l5_preds)))
print("R^2: ", r2_score(y_valid, l5_preds))
```

```python
from yellowbrick.regressor import PredictionError, ResidualsPlot
visualizer = PredictionError(l5)
visualizer.fit(x_train, y_train)  # Fit the training data to the visualizer
visualizer.score(x_valid, y_valid)  # Evaluate the model on the test data
visualizer.poof() 
 
```
